```python
import pandas as pd
import matplotlib.pyplot as plt

# Load results
results = pd.read_csv('../results/metrics.csv')

print("="*60)
print("FINAL RESULTS SUMMARY")
print("="*60)
print(results)

# Calculate performance degradation
print("\n" + "="*60)
print("PERFORMANCE DEGRADATION ANALYSIS")
print("="*60)

clean_cf_precision = results['cf_precision'].iloc[0]
clean_gan_precision = results['gan_precision'].iloc[0]

for i in range(1, len(results)):
    noise_level = results['test_set'].iloc[i]
    cf_drop = ((clean_cf_precision - results['cf_precision'].iloc[i]) / clean_cf_precision) * 100
    gan_drop = ((clean_gan_precision - results['gan_precision'].iloc[i]) / clean_gan_precision) * 100
    
    print(f"\n{noise_level}:")
    print(f"  CF dropped: {cf_drop:.2f}%")
    print(f"  GAN dropped: {gan_drop:.2f}%")
    print(f"  Winner: {'GAN' if gan_drop < cf_drop else 'CF'} (more robust)")
    
```
